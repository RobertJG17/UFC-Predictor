```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master = pd.read_csv('../Training Data/ufc-master.csv')
```

```python
master = master.iloc[12:, :].reset_index().drop("index", axis=1)
```

```python
master.head(3)
```

```python
# INITIAL FEATURE SELECTION: 
# odds, ev, dec_odds, sub_odds, ko_odds, current_win_streak, current_lose_streak,
```

```python
for col in master.columns:
    print(col)
```

```python
selected_features = master[["R_fighter", "R_odds", "R_ev", "R_avg_SIG_STR_pct", "R_avg_TD_pct", "R_current_win_streak",
                            "r_ko_odds", "r_sub_odds",
                            
                            "B_fighter", "B_odds", "B_ev", "B_avg_SIG_STR_pct", "B_avg_TD_pct", "B_current_win_streak",
                            "b_ko_odds", "b_sub_odds",
                            "Winner"]]
```

```python
selected_features["Winner"] = selected_features.Winner.map({"Red": 1, "Blue": 0})
```

```python
selected_features.Winner
```

```python
# MOVE TO SEPARATE EDA NOTEBOOK

plt.figure(figsize=(16, 8), dpi=200)
sns.heatmap(selected_features.corr(), annot=True)
plt.show();
```

```python
# MOVE TO SEPARATE ETL NOTEBOOK

nan_ser = selected_features.isnull().sum() / len(selected_features)
nan_ser.sort_values()
```

```python
# MOVE TO SEPARATE ETL NOTEBOOK

selected_features.dropna(subset=["r_ko_odds", "r_sub_odds",
                                 "R_avg_TD_pct", "R_avg_SIG_STR_pct",
                                 "b_sub_odds", "b_ko_odds",
                                 "B_avg_TD_pct", "B_avg_SIG_STR_pct"], axis=0, inplace=True)
selected_features.reset_index(inplace=True)
```

```python
# MOVE TO SEPARATE ETL NOTEBOOK

selected_features.drop("index", axis=1, inplace=True)
```

```python
selected_features
```

```python
X = selected_features.drop(["R_fighter", "B_fighter", "Winner"], axis=1)
```

```python
y = selected_features["Winner"].map({1: "Red", 0: "Blue"})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=17)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_val = scaler.transform(X_val)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
svc = SVC(class_weight="balanced", probability=True)
```

```python
param_grid = {
    "C": np.linspace(0.01, 2, 11),
    "gamma": ["scale", "auto"],
    "kernel": ["rbf", "poly", "sigmoid"]
}
```

```python
grid_model = GridSearchCV(svc,
                          param_grid=param_grid)
```

```python
grid_model.fit(scaled_X_train, y_train)
```

```python
grid_model.best_params_
```

```python
predictions = grid_model.predict(scaled_X_val)
```

```python
probabilities = grid_model.predict_proba(scaled_X_val)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=200)
plot_confusion_matrix(grid_model, scaled_X_val, y_val, ax=ax)
```

```python
print(classification_report(y_val, predictions))
```
