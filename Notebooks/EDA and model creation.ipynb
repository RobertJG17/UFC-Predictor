```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
master = pd.read_csv('../Training Data/ufc-master.csv')
```

```python
# TAKE THE DATE OF THE FIRST ENTRY AND EXCLUDE ALL ENTRIES WITH THAT DATE (EXCLUDES MOST RECENT EVENT)

master = master.loc[master.date != master.iloc[0]['date'], :].reset_index().drop("index", axis=1)
```

```python
master.head(3)
```

```python
plt.figure(figsize=(16,8), dpi=200)
sns.countplot(data=master, x="Winner");
```

```python
# NUM OF ROWS IN THE DATASET

len(master)
```

```python
# WINNER -> {"Red": 1, "Blue": 0}

master["Winner Dummies"] = master.Winner.map({"Red": 1, "Blue": 0})
```

```python
# TOP 20 FEATURES THAT ARE MOST HIGHLY CORRELATED WITH WINNING (BY ABSOLUATE VAL)

pearson_corr = abs(master.corr()["Winner Dummies"]).sort_values(ascending=False)
```

```python
plt.figure(figsize=(16,8), dpi=200)

sns.barplot(y=pearson_corr.index[1:21], x=pearson_corr.values[1:21], orient='h')
plt.title("Feature Correlations to Winning by Pearson Ranking")

fig.tight_layout()
```

```python
# WE WILL CONSIDER THE HIGHEST CORRELATED FEATURES THAT ARE NOT UNIQUE TO WEIGHTCLASSES OR GENDERS

features = pearson_corr.loc[
    [
        "B_odds", "R_odds", 
        "B_ev", "R_ev", 
        "b_dec_odds", "r_dec_odds",
        "b_ko_odds", "r_ko_odds", 
        "b_sub_odds", "r_sub_odds",
    ]
]
```

```python
nan_ser = master.loc[:, features.index].isnull().sum() / len(master)

nan_ser.sort_values()
```

```python
X = master.dropna(subset=nan_ser.sort_values().index, axis=0).loc[:, features.index]
```

```python
# NUM OF ROWS IN CLEANED DATAFRAME WITH SELECTED FEATURES

len(X)
```

```python
y = master.dropna(subset=nan_ser.sort_values().index, axis=0).loc[:, "Winner"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=17)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
```

```python
# CREATE RANDOM FOREST CLASSIFIER MODEL

rfc = RandomForestClassifier()
```

```python
# FOR SVC MODEL

# param_grid = {
#     "C": np.arange(0.1, 1.1, 0.1),
#     "kernel": ['rbf', 'poly', 'sigmoid'],
#     "class_weight": ["balanced"]
# }

# FOR RFC ENSEMBLE MODEL

param_grid = {
    "max_depth": [10, 12, 14, 16],
    "n_estimators": [100, 110, 120, 130],
    "class_weight": ["balanced", "balanced_subsample"]
}
```

```python
# USE GRIDSEARCH TO FIND OPTIMAL HYPERPARAMETERS

grid = GridSearchCV(estimator=rfc, param_grid=param_grid)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
model = grid.best_estimator_
```

```python
model.fit(X_train, y_train)
```

```python
# GRAB FEATURE IMPORTANCES FROM OUR ADABOOSTER AND USE A LINECHART TO VISUALIZE THEM

feat_imp = model.feature_importances_
feature_labels = X.columns
feat_ser = pd.Series(data=feat_imp, index=feature_labels).sort_values()
```

```python
import plotly.express as px
```

```python
px.line(x=feat_ser.index, y=feat_ser.values, title="Feature vs. Feature Importance")
```

```python
predictions = model.predict(X_val)
```

```python
probabilities = model.predict_proba(X_val)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=200)
plot_confusion_matrix(model, X_val, y_val, ax=ax)
```

```python
print(classification_report(y_val, predictions))
```
